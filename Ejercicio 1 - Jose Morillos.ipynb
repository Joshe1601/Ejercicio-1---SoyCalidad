# Ejercicio 1:

## Enunciado

Se tienen los siguientes diccionarios:

```python
Productos = {1:'Pantalones', 2:'Camisas', 3:'Corbatas', 4:'Casacas'}
Precios = {1:200.00, 2:120.00, 3:50.00, 4:350.00}
Stock = {1:50, 2:45, 3:30, 4:15}
```

Elaborar un programa que muestre los diccionarios, y programar las siguientes acciones:

[1] Agregar

[2] Eliminar

[3] Actualizar

[4] Salir

El resultado debe mostrar lo siguiente:

========================================

Lista de Productos:

========================================

1 	 Pantalones	 200.0 	 50

2 	 Camisas 	 120.0 	 45

3 	 Corbatas 	 50.0 	 30

4 	 Casacas 	 350.0 	 15

========================================

[1] Agregar, [2] Eliminar, [3] Actualizar, [4] Salir

Elija opción:

## Desarrollo

```python
import time
# Funciones de validación
def validate_cost(cost_str):
    try:
        cost = float(cost_str)
        if cost <= 0:
            print("El precio debe ser mayor que 0")
            return None
        return cost
    except ValueError:
        print("Por favor, ingrese un número válido para el precio")
        return None

def validate_stock(stock_str):
    try:
        stock = int(stock_str)
        if stock < 0:
            print("El stock no puede ser negativo")
            return None
        return stock
    except ValueError:
        print("Por favor, ingrese un número entero válido para el stock")
        return None

def validate_name(name):
    if name.strip() == "":
        print("El nombre no puede estar vacío")
        return None
    if len(name) > 20:
        print("El nombre no puede tener más de 20 caracteres")
        return None
    return name.strip()
```

```python
# Funciones del menú
def show_products():
    print("=" * 40)
    print("Lista de Productos:")
    print("=" * 40)
    print(f"{'ID':<10}{'Producto':<14}{'Precio':<10}{'Stock'}")
    print("-" * 40)
    for id in Productos:
        print(f"{id:<10}{Productos[id]:<14}{Precios[id]:<10}{Stock[id]}")
    print("=" * 40)

def add_products():
    try:
        id = max(Productos.keys()) + 1
        
        while True:
            name = validate_name(input("Ingrese nombre del producto: "))
            if name is not None:
                break
        
        while True:
            cost = validate_cost(input("Ingrese precio del producto: "))
            if cost is not None:
                break
        
        while True:
            stock = validate_stock(input("Ingrese stock del producto: "))
            if stock is not None:
                break
        
        Productos[id] = name
        Precios[id] = cost
        Stock[id] = stock
        print("\nProducto agregado exitosamente\n\n")
    except Exception as e:
        print(f"\nError al agregar producto: {str(e)}\n\n")

def delete_products():
    try:
        if not Productos:
            print("\nNo hay productos para eliminar")
            return
        
        while True:
            try:
                id = int(input("Ingrese ID del producto a eliminar: "))
                if id in Productos:
                    del Productos[id]
                    del Precios[id]
                    del Stock[id]
                    print("\nProducto eliminado exitosamente.\n\n")
                    break
                else:
                    print("\nID de producto no encontrado.\n\n")
            except ValueError:
                print("\nPor favor, ingrese un número válido para el ID\n\n")
    except Exception as e:
        print(f"\nError al eliminar producto: {str(e)}\n\n")

def update_products():
    try:
        if not Productos:
            print("\nNo hay productos para actualizar\n\n")
            return
        
        while True:
            try:
                id = int(input("Ingrese ID del producto a actualizar: "))
                if id in Productos:
                    while True:
                        nombre = validate_name(input("Nuevo nombre del producto: "))
                        if nombre is not None:
                            break
                    
                    while True:
                        precio = validate_cost(input("Nuevo precio del producto: "))
                        if precio is not None:
                            break
                    
                    while True:
                        stock = validate_stock(input("Nuevo stock del producto: "))
                        if stock is not None:
                            break
                    
                    Productos[id] = nombre
                    Precios[id] = precio
                    Stock[id] = stock
                    print("\nProducto actualizado exitosamente.\n\n")
                    break
                else:
                    print("\nID de producto no encontrado.\n\n")
            except ValueError:
                print("\nPor favor, ingrese un número válido para el ID\n\n")
    except Exception as e:
        print(f"\nError al actualizar producto: {str(e)}\n\n")
```

```python
# Programa principal
def main():
    print("\nBienvenido")
    while True:
        try:
            show_products()
            print("\n[1] Agregar, [2] Eliminar, [3] Actualizar, [4] Salir")
            time.sleep(1.5)
            option = input("\nElija opción: ")
            print(f"Opción escogida: {option}")
            
            match option:
                case "1":
                    add_products()
                case "2":
                    delete_products()
                case "3":
                    update_products()
                case "4":
                    print("\n¡Gracias por usar el programa! Hasta pronto :)")
                    break
                case _:
                    print("\nOpción inválida. Intente de nuevo con otra opción.")
        except Exception as e:
            print(f"\nError inesperado: {str(e)}")

if __name__ == "__main__":
    main()
```
